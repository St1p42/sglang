```python
# Compare the saved results side by side
import json
from pathlib import Path

import pandas as pd

base = Path("/content/sglang/690AB/eval/experiment_1/results")
rows = []
for name in ["sglang_multiturn.jsonl", "vllm_multiturn.jsonl"]:
    path = base / name
    with path.open() as f:
        rows.append(json.loads(f.readlines()[-1]))

df = pd.DataFrame(rows)
df["requests_per_sec"] = df["num_requests"] / df["latency"]
df[["backend", "latency", "num_requests", "num_turns", "requests_per_sec", "other"]]
```
