```python
# Run the latency-style DSPy RAG benchmark 5 times against vLLM
%%bash
set -e
cd /content/sglang/690AB/eval/experiment_1/rag_dspy_style
mkdir -p results
rm -f results/vllm_latency.jsonl results/vllm_latency_run*.json
for run_id in 1 2 3 4 5; do
curl -s http://127.0.0.1:21000/generate -H "Content-Type: application/json" -d '{"prompt":"Context:\\nDummy context for vLLM warmup.\\n\\nQuestion: What is 1 plus 1?\\nAnswer:","temperature":0.0,"max_tokens":16,"n":1}'
  TF_CPP_MIN_LOG_LEVEL=3 python bench_dspy_rag_serving.py --backend vllm --model "$MODEL" --dataset-file "$WORKLOAD_FILE" --max-requests "$LATENCY_REQUESTS" --parallel "$LATENCY_PARALLEL" --result-file results/vllm_latency.jsonl --raw-result-file results/vllm_latency_run${run_id}.json --run-id ${run_id}
done
```

```python
# Aggregate the latency-style vLLM runs
import glob, json, pandas as pd

rows = []
for path in sorted(glob.glob("/content/sglang/690AB/eval/experiment_1/rag_dspy_style/results/vllm_latency_run*.json")):
    with open(path) as f:
        obj = json.load(f)
    rows.append({
        "run_id": obj.get("run_id"),
        "mean_e2e_latency_ms": obj["metrics"]["mean_e2e_latency_ms"],
        "mean_ttft_ms": obj["metrics"]["mean_ttft_ms"],
        "p95_e2e_latency_ms": obj["extended_metrics"].get("p95_e2e_latency_ms"),
        "request_throughput": obj["metrics"]["request_throughput"],
        "cache_hit_rate": obj.get("cache_hit_rate"),
        "rouge_l": obj.get("answer_quality", {}).get("rouge_l"),
        "token_f1": obj.get("answer_quality", {}).get("token_f1"),
    })
df = pd.DataFrame(rows)
display(df)
summary = df.drop(columns=["run_id"]).agg(["mean", "std"]).T.reset_index().rename(columns={"index": "metric"})
display(summary)
```
