```python
import json
import math
import subprocess
import sys
import time
from pathlib import Path

import requests

# Files (4B = HiCache OFF)
RESULTS_FILE = Path("/content/sglang/690AB/Experiment4/multiturnhicache_benchmark_4b_strong.txt")
JSONL_FILE = Path("/content/sglang/690AB/Experiment4/multiturnhicache_benchmark_4b_strong_runs.jsonl")

RUNS = 5
HOST = "127.0.0.1"
PORT = 30000
FLUSH_CACHE_URL = f"http://{HOST}:{PORT}/flush_cache"

METRICS = [
    ("total_requests", "Total requests"),
    ("request_rate", "Request rate"),
    ("average_prompt_len", "Average Prompt Length"),
    ("average_output_len", "Average Output Length"),
    ("average_ttft", "Average TTFT"),
    ("p90_ttft", "P90 TTFT"),
    ("median_ttft", "Median TTFT"),
    ("average_latency", "Average latency"),
    ("p90_latency", "P90 latency"),
    ("median_latency", "Median latency"),
    ("input_token_throughput", "Input token throughput"),
    ("output_token_throughput", "Output token throughput"),
    ("throughput", "Request Throughput"),
    ("cache_hit_rate", "Cache Hit Rate"),
]

def mean(xs):
    return sum(xs) / len(xs) if xs else 0.0

def std(xs):
    if not xs:
        return 0.0
    m = mean(xs)
    return math.sqrt(sum((x - m) ** 2 for x in xs) / len(xs))

def fmt(v):
    if isinstance(v, int):
        return str(v)
    return f"{v:.6f}"

def read_last_jsonl_row(path: Path):
    lines = path.read_text().strip().splitlines()
    if not lines:
        raise RuntimeError(f"No rows found in {path}")
    return json.loads(lines[-1])

RESULTS_FILE.parent.mkdir(parents=True, exist_ok=True)
if JSONL_FILE.exists():
    JSONL_FILE.unlink()

all_summaries = []
lines = []
lines.append("Multiturn HiCache Benchmark 4B (HiCache OFF - Final Strong Workload)")
lines.append("")

for run_idx in range(1, RUNS + 1):
    print(f"Starting Run{run_idx}...")
    print(f"Flushing cache before Run{run_idx}...")

    resp = requests.post(FLUSH_CACHE_URL)
    resp.raise_for_status()
    time.sleep(1)

    cmd = [
        sys.executable,
        "-u",
        "benchmark/hicache/bench_multiturn.py",
        "--model-path", "Qwen/Qwen2.5-7B-Instruct",
        "--host", HOST,
        "--port", str(PORT),
        "--disable-auto-run",
        "--request-rate", "6",
        "--distribution", "uniform",
        "--seed", "1",
        "--num-clients", "32",
        "--max-parallel", "16",
        "--num-rounds", "4",
        "--request-length", "2048",
        "--sub-question-input-length", "512",
        "--output-length", "64",
        "--log-file", str(JSONL_FILE),
        "--tag", f"Run{run_idx}",
    ]

    subprocess.run(cmd, check=True, cwd="/content/sglang")

    row = read_last_jsonl_row(JSONL_FILE)
    summary = row["summary"]
    all_summaries.append(summary)

    lines.append(f"Run{run_idx}")
    for key, label in METRICS:
        lines.append(f"{label}: {fmt(summary[key])}")
    lines.append("")

lines.append("Mean/Std")
for key, label in METRICS:
    values = [summary[key] for summary in all_summaries]
    lines.append(f"{label}: mean={mean(values):.6f}, std={std(values):.6f}")

RESULTS_FILE.write_text("\n".join(lines))

print(f"\nSaved results to: {RESULTS_FILE}")
print(RESULTS_FILE.read_text())
```
